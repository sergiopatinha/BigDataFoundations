```python

#Data Cleaning

import duckdb, pandas as pd
from pathlib import Path

# create or connect if it already exists
con = duckdb.connect("movielens100K.duckdb")
```

```python
#creation of a view with movies that have at least 10 ratings

con.sql("""
CREATE OR REPLACE VIEW ratings_clean AS
SELECT *
FROM ratings
WHERE movieId IN (
  SELECT movieId
  FROM ratings
  GROUP BY movieId
  HAVING COUNT(*) >= 10
)
""")
```
