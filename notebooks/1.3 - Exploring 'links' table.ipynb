```python

#initial exploration of "links" table

import duckdb, pandas as pd
from pathlib import Path

#create or connect if it already exists
con = duckdb.connect("movielens100K.duckdb")
```

### initial exploration of the table "links"


 - To know Before:
 - movieId is an identifier for movies used by <https://movielens.org>. E.g., the movie Toy Story has the link <https://movielens.org/movies/1>

 - imdbId is an identifier for movies used by <http://www.imdb.com>. E.g., the movie Toy Story has the link --<http://www.imdb.com/title/tt0114709/>.

 - tmdbId is an identifier for movies used by <https://www.themoviedb.org>. E.g., the movie Toy Story has the link --<https://www.themoviedb.org/movie/862>.

 - the use of the resources listed above is subject to the terms of each provider.


```python
#check the table structure (columns and data types)
con.sql("DESCRIBE links").df()
```

####  Comment

 - "movieID":  BIGINT;
 - "imdbID": VARCHAR
 - "tmdbID":  BIGINT;

```python
#Ver o tipo de dados de cada coluna
con.sql("PRAGMA table_info('links')").df()
```

#### Comments

- The column `notnull` (boolean) shows four zero values.  
  - This means the table allows null (NULL) values.  
  - Since the DataFrame was loaded with `read_csv_auto`, DuckDB does not apply NOT NULL constraints.

- The column `dflt_value` shows the default value assigned when no other value is specified during an INSERT.  
  - In datasets imported with `read_csv_auto` (such as MovieLens), this column almost always appears as NULL.

- The column `pk` (boolean) indicates whether the column is part of the table’s primary key.  
  - It is not.


```python
#check first 10 rows
con.sql("SELECT * FROM links LIMIT 10").df()
```

```python
#count total number of rows
con.sql("SELECT COUNT(*) AS total_registos FROM links").df()
```

```python
#count the number of missing values
con.sql("""
SELECT
    COUNT(*) - COUNT(movieId) AS missing_movieId,
    COUNT(*) - COUNT(imdbID) AS missing_imdbID,
    COUNT(*) - COUNT(tmdbID) AS missing_tmdbID
FROM links
""").df()
```
