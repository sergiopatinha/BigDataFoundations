```python
#================================================
# LIMPEZA DE DADOS
#================================================

import duckdb, pandas as pd
from pathlib import Path

# cria ou liga se já existir:
con = duckdb.connect("movielens100K.duckdb")
```

```python
#Criação de uma view com os filmes que têm pelo menos 10 ratings

con.sql("""
CREATE OR REPLACE VIEW ratings_clean AS
SELECT *
FROM ratings
WHERE movieId IN (
  SELECT movieId
  FROM ratings
  GROUP BY movieId
  HAVING COUNT(*) >= 10
)
""")
```
